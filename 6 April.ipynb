### Support Vector Classifier

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# create a synthetic dataset 
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_classes=2,n_clusters_per_class=2,n_features=2,n_redundant=0)
```

```python
x
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

### train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(x_train,y_train)
```

```python
svc.coef_  # it gives coefficient 1 and 2 (coefficient is slope)
```

```python
y_pred=svc.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))  # 89% accuracy
```

### Hyperparameter tunning with svc

```python
from sklearn.model_selection import GridSearchCV
param_grid={'C':[0.1,1,10,100,1000],
            'gamma':[1,0.1,0.01,0.001,0.0001],
            'kernel':['linear']
}
```

```python
# verbose will gives all the values we are getting in a sequence
grid=GridSearchCV(svc,param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
y_pred2=grid.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred2))
print(accuracy_score(y_test,y_pred2))
print(classification_report(y_test,y_pred2))   # it is giving 90% accuracy
```
